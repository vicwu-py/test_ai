```python
def menu(wine,entree,dessert):
    print(f"酒類:{wine}")
    print(f"主食：{entree}")
    print(f"甜點:{dessert}")
```

```python
#引數位置
#依據參數的順序

menu('白酒','牛排','蛋糕')
```

```python
#引數名稱呼叫
#可以不依照順序
menu(entree='牛排',dessert='蛋糕',wine='白酒')
```

```python
#混合呼叫
#前面必需是引數位置呼叫
#後面是引數名稱呼叫
#使用引數名稱呼叫後，後面全部都要使用引數名稱呼叫

menu('白酒',dessert='蛋糕',entree='牛排')
```

```python
#參數有default值
#必需放在最後面

def  menu(wine,entree,dessert='奶昔'):
    print(f"酒類={wine}")
    print(f"主食={entree}")
    print(f"甜點={dessert}")
```

```python
menu('白酒','牛排')
```

```python
menu('白酒','牛排',dessert='蛋糕')
```

```python
#有*的參數使用
#呼叫時沒有限定引數位置呼叫的數量
def print_args(*args):
    print(args)
    print("========")

print_args()
print_args(1,2,3,4,'a','b','true')
```

```python
#有**的參數
#沒有限定數量的引數名稱呼叫

def print_kwargs(**kwargs):
    print(kwargs)
    print("===========")    

print_kwargs(one=1,two=2,three=3,four=4,five=5)
print_kwargs()
```
